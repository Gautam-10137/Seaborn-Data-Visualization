 ## Distribution plot  express one continuous  feauture

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(4,2),dpi=200)
sns.rugplot(x='salary',data=df,height=0.5)  # here y axis values are not meaningful
#  each single tick represent a single salary
```

```python
# .distplot -> deprecated
sns.displot(x="salary",data=df)
```

```python
#  customizing no. of bins
sns.displot(x="salary",data=df,bins=50)
```

```python
#  customizing bakground
# sns.set(style='darkgrid') # -> .set(style) is a high level command -> set style for rest of the notebook 
sns.displot(x="salary",data=df) # -> all matplotlib customization are available in this block
```

```python
sns.displot(x="salary",data=df,color='purple',edgecolor='pink',linewidth=2,ls="--")
```

## kernel density estimation and rug in displot

```python
sns.displot(x="salary",data=df,kde=True,rug=True)
```

```python
#  displot is more generalized version of histplot
sns.histplot(x='salary',data=df)
```

```python
#  kde plot
sns.kdeplot(x='salary',data=df)
```

```python
#  performing kde distribution in uniformly random dataset
np.random.seed(42)
sample_ages=np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages=pd.DataFrame(sample_ages,columns=['age'])
sample_ages
```

```python
sns.rugplot(x='age',data=sample_ages)
```

```python
sns.displot(x='age',data=sample_ages,rug=True,bins=30,kde=True)
# here kde curve is cut off at 0 & 100 (i.e having actual realistic values)
```

```python
#  but by default kde curve is
sns.kdeplot(x="age",data=sample_ages) # also include non realistic values around -ve 
# to cutoff this we have -> clip=[range]
```

```python
sns.kdeplot(x="age",data=sample_ages,clip=[0,100])
```

```python
#  adjusting bandwidth -> higher the bandwidth - higher the smoothness
```

```python
sns.kdeplot(x="age",data=sample_ages,clip=[0,100],bw_adjust=0.5)
```

```python
#  shading the area
sns.kdeplot(x="age",data=sample_ages,clip=[0,100],bw_adjust=0.4,fill=True)
```
