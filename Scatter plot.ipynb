```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing seaborn library
import seaborn as sns
```

```python
df=pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
# syntax
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
#  also can customize in same cell
plt.figure(figsize=(14,5),dpi=300)
sns.scatterplot(x='salary',y='sales',data=df)
# Note -> seaborn is calling matplotlib at backend
```

## parameters

```python
#  coloring the points based on a particular column -> hue()
plt.figure(figsize=(14,5),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education') # hueing a categorical column
```

```python
#  hueing a continuous column
plt.figure(figsize=(14,5),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='salary')
```

```python
#  customizing the color acc. to colormap in matplotlib
plt.figure(figsize=(14,5),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2')
```

```python
#  size of points based on a particular column
plt.figure(figsize=(14,5),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,size='work experience')
```

```python
#  changing general size -> s=
plt.figure(figsize=(14,5),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,s=500)
```

```python
#  distinguishing overlapping points by controlling transparency -> alpha=
plt.figure(figsize=(14,5),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,s=500,alpha=0.2)
```

```python
#  styling based on a particular column rather than coloring ->style
plt.figure(figsize=(6,3),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,s=200,style='level of education',alpha=0.7,hue='level of education')

#  saving
plt.savefig("ColorfulPlot.jpg",bbox_inches='tight')
```
