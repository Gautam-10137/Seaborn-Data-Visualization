```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
#  catplot -> backgroud call of categorical plots e.g. box plot
sns.catplot(data=df,x='math score',y='gender',kind='box')  # -> box plot
```

```python
#  making grid based on a particular feature
sns.catplot(data=df,x='math score',y='gender',kind='box',row='lunch') # or as we did hue to separate based on feature 
```

```python
sns.catplot(data=df,x='math score',y='gender',kind='box',row='lunch',col='test preparation course')
```

```python
#  PairGrid
g=sns.PairGrid(data=df)
g=g.map_upper(sns.scatterplot)
g=g.map_diag(sns.histplot)
g=g.map_lower(sns.kdeplot)
```
