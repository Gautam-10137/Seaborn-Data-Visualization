```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,3))
sns.countplot(x='division',data=df)
# also can set y limits
plt.ylim(90,260)   # -> more clarity
```

```python
plt.figure(figsize=(10,3))
sns.countplot(x='level of education',data=df)
```

```python
#  also can add hue paramter to show more information
plt.figure(figsize=(10,3),dpi=200)
sns.countplot(x='level of education',data=df,hue='division')
```

```python
#  barplot
plt.figure(figsize=(10,4))
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,errorbar='sd')
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,errorbar='sd',hue='division')
# positioning the legend
plt.legend(bbox_to_anchor=(1.05,1))
```

## Distribution within categories

```python
#  comparing ditribution across categories
```

```python
df=pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
# boxplot
sns.boxplot(y='math score',x='race/ethnicity',data=df);
```

```python
#  creating subcategories
plt.figure(figsize=(10,4))
sns.boxplot(y='math score',x='parental level of education',data=df,hue="test preparation course");
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
#  creating horizontally
plt.figure(figsize=(10,6))
sns.boxplot(x='math score',y='parental level of education',data=df,hue="test preparation course",palette="Set2");
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
# violin plot
plt.figure(figsize=(10,6))
sns.violinplot(x='math score',y='parental level of education',data=df);
```

```python
#  adding subcategories in violin plot
plt.figure(figsize=(10,6))
sns.violinplot(x='math score',y='parental level of education',data=df,hue="test preparation course",palette="Set2");
plt.legend(bbox_to_anchor=(1.25,1))
```

```python
#  to avoid stickness of subcategories -> split=True
plt.figure(figsize=(10,6))
sns.violinplot(x='math score',y='parental level of education',data=df,hue="test preparation course",palette="Set2",split=True);
plt.legend(bbox_to_anchor=(1.25,1))
```

```python
#  to remove inner boxe => inner=None
#  also can inner='quartile'
plt.figure(figsize=(10,6))
sns.violinplot(x='math score',y='parental level of education',data=df,inner='quartile');
```

```python
#  to pick up more noise i.e. decreasing bandwith of gaussian distribution
plt.figure(figsize=(10,6))
sns.violinplot(x='math score',y='parental level of education',data=df,bw=0.1);
 
```

```python
#  swarm plot
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(x='math score',data=df,size=3)
```

```python
#  boxenplot
plt.figure(figsize=(10,6))
sns.boxenplot(x='math score',y='parental level of education',data=df);
```
