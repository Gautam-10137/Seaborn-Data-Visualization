```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
#  joint plot
sns.jointplot(x='reading score',y='math score',data=df,kind='scatter',alpha=0.4)
```

```python
sns.jointplot(x='reading score',y='math score',data=df,kind='hist',alpha=0.8)
```

```python
#  kde
sns.jointplot(x='reading score',y='math score',data=df,kind='kde',fill=True)
```

```python
# pair plot
sns.pairplot(data=df,hue='gender')
```

```python
#  to remove duplicates
sns.pairplot(data=df,hue='gender',corner=True)
```
